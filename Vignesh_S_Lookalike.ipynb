```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
df_transaction = pd.read_csv('df_transaction.csv') #loading the transaction dataset in which EDA is done
```

```python
df_transaction.head()
```

```python
pd.options.display.max_rows = None
pd.options.display.max_columns = None
pd.options.display.max_colwidth = None
```

```python
# Create a matrix where rows represent customers and columns represent products with quantities as values
customer_product_matrix = df_transaction.pivot_table(index='CustomerID', columns='ProductID', values='Quantity', fill_value=0)
customer_product_matrix
```

```python
# Scale the matrix to standardize the data
scaler = StandardScaler()
scaled_matrix = scaler.fit_transform(customer_product_matrix)
scaled_matrix
```

```python
# Calculate cosine similarity between customers
cosine_sim_matrix = cosine_similarity(scaled_matrix)
df_similarity = pd.DataFrame(cosine_sim_matrix, index = customer_product_matrix.index, columns = customer_product_matrix.index)
df_similarity
```

```python
top_20 = [f'C{str(i).zfill(4)}' for i in range(1, 21)]
```

```python
lookalikes = {}
for cus in top_20:
    similar_customers = df_similarity[cus].sort_values(ascending=False)[1:4]
    # Store results as a list of tuples (CustomerID, similarity score)
    lookalikes[cus] = list(zip(similar_customers.index, similar_customers.values))

lookalikes
```

```python
# Convert results into a DataFrame and save as CSV
df_lookalikes = pd.DataFrame([{'CustomerID': customer, 'Lookalikes': lookalikes[customer]} for customer in lookalikes])
df_lookalikes.to_csv('Vignesh_S_Lookalike.csv', index=False)
df_lookalikes
```
